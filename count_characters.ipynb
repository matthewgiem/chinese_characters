```python
import glob
from matplotlib import pyplot as plt
import numpy as np
```

```python
dictionary = dict()
num_of_pics = []
name_of_pics = []
for folder in glob.glob('ignore_folder/data/*'):
    count = 0
    for file in glob.glob('{}/*'.format(folder)):
#         print(file.split('/')[2])
        count += 1
#     print('{} has {}'.format(folder.split('/')[1], count))
    dictionary[folder.split('/')[1]] = count
    num_of_pics.append((folder.split('/')[2], count))
    name_of_pics.append(folder.split('/')[2])
```
